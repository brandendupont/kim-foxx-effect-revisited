```python
import pandas as pd
```

```python
#unemployment data from Bureau of Labor Statistics
#https://beta.bls.gov/dataQuery/find?fq=survey:%5Bla%5D&s=popularity:D&q=cook+county
ue = pd.read_csv(r'C:\Users\Gabriel\OneDrive\CUSP\Fall 2021 Classes\PUI\Final Project\kim-foxx-effect-revisited\data\unemployment_data_2000-2018.csv')
ue.head()
```

```python
#slice month string and reformat as int
ue['month'] = ue.Period.apply(lambda x: int(x[1:]))
```

```python
ue['month_dt'] = pd.to_datetime({'year':ue.Year, 'month':ue.month, 'day':1})
ue.head()
```

```python
#save to file
ue.to_csv(r'C:\Users\Gabriel\OneDrive\CUSP\Fall 2021 Classes\PUI\Final Project\kim-foxx-effect-revisited\data\unemployment_data_tomerge.csv')
```

```python
#weather data from NOAA Climate Data Online Search
# https://www.ncdc.noaa.gov/cdo-web/search
weather = pd.read_csv(r'C:\Users\Gabriel\OneDrive\CUSP\Fall 2021 Classes\PUI\Final Project\kim-foxx-effect-revisited\data\weather_data_2000-2018.csv')
weather.head()
```

```python
weather['year'] = weather.DATE.apply(lambda x: int(x[:4]))
weather['month'] = weather.DATE.apply(lambda x: int(x[5:]))
weather['month_dt'] = pd.to_datetime({'year':weather.year,'month':weather.month,'day':1})
weather.head()
```
